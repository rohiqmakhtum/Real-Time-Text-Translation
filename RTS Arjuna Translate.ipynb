```python
import mysql.connector
import pandas as pd

db_params = {
    "host": "arjuna.genah.net",
    "user": "arjuna_read_only",
    "port": 33061,
    "password": "AcgK_GWB-24",
    "database": "arjuna"
}

try:
    db_connection = mysql.connector.connect(**db_params)

    # Menyiapkan query SQL
    query = """
    SELECT
    * FROM customer_rts_histories WHERE (returned_at BETWEEN '2024-08-01' AND '2024-08-31')
    """

    df = pd.read_sql(query, db_connection)
    print(df)
except mysql.connector.Error as err:
    print(f"Error: {err}")
finally:
    # Menutup koneksi
    if db_connection.is_connected():
        db_connection.close()
```

```python
# Memfilter data berdasarkan kolom 'reason' yang mengandung 'Return to sender%'
df_rts = df[df['reason'].str.contains("Return to sender", case=False, na=False)]
df_rts
```

```python
from googletrans import Translator
translator = Translator()

print("Sebelum diterjemahkan:", df_rts['reason'].head())

df_rts['reason_translated'] = df_rts['reason'].apply(lambda x: translator.translate(x, src='en', dest='id').text if isinstance(x, str) else x)
print("Setelah diterjemahkan:", df_rts[['reason', 'reason_translated']].head())
```
